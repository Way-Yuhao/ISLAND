## Alternative broadband emissivity calculations

Previously from notebook **12 - Surfrad**, we are using Ogawa's equation to calculate broadband emissivity at surfrad stations, and got: </br>
    BND station broadband emissivity = 0.9751</br>
    TBL station broadband emissivity = 0.9711</br>
    DRA station broadband emissivity = 0.9698</br>
    FPK station broadband emissivity = 0.9791</br>
    GWN station broadband emissivity = 0.9690</br>
    PSU station broadband emissivity = 0.9717</br>
    SXF station broadband emissivity = 0.9727</br>
    SGP station broadband emissivity = 0.9743</br>

```python
from eval.surfrad import get_emis_at
from omegaconf import DictConfig, OmegaConf

geemap.Map()
config = OmegaConf.load('../config/surfrad.yaml')
for station in config.stations:
    lon = config['stations'][station]['Longitude']
    lat = config['stations'][station]['Latitude']
    ogawa_emis = get_emis_at(lon, lat, use_equation='ogawa')
    emis = get_emis_at(lon, lat, use_equation='cheng')
    print(f'{station} station broadband emissivity = {emis:.4f} ({(emis - ogawa_emis):.4f})')
```

```python
station_id = 'BND'
start_date = '20130411'
end_date = '20201231'
region_dir = f'/home/yuhaoliu/Data/ISLAND/surfrad_val/{station_id}'
df = load_datapoints(station_id, start_date, end_date, region_dir, use_emis_from='cheng')
```

```python
plot_regression(df, 'surfrad_lst', 'surfrad_air_temp', 'all', f'Discrepency between air temp and surface temp ({station_id})')
plot_regression(df, 'surfrad_lst', 'landsat_lst', 'clear', f'Cloud-free LST at {station_id}')
plot_regression(df, 'surfrad_lst', 'island_lst', 'cloudy', f'Cloud-occluded LST at {station_id}')
```

```python
station_id = 'TBL'
start_date = '20130411'
end_date = '20201231'
region_dir = f'/home/yuhaoliu/Data/ISLAND/surfrad_val/{station_id}'
df = load_datapoints(station_id, start_date, end_date, region_dir, use_emis_from='cheng')
plot_regression(df, 'surfrad_lst', 'surfrad_air_temp', 'all', f'Discrepency between air temp and surface temp ({station_id})')
plot_regression(df, 'surfrad_lst', 'landsat_lst', 'clear', f'Cloud-free LST at {station_id}')
plot_regression(df, 'surfrad_lst', 'island_lst', 'cloudy', f'Cloud-occluded LST at {station_id}')
```
